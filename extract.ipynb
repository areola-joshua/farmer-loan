```python
df = pd.read_csv("FarmingHouseholds_ano.csv")

# Sample 30% of each state group
df_sampled = (
    df.groupby('state_code', group_keys=False)
    .apply(lambda x: x.sample(frac=0.3, random_state=42))
)

df_sampled.to_csv("FarmingHouseholds_30percent_sampled.csv", index=False)
```

```python
columns_to_keep = {
    'HHID': 'household_id',
    'state_code': 'state',
    'lga_code': 'lga',
    'hh_sex': 'hh_head_sex',
    'hh_age': 'hh_head_age',
    'hh_school_attd': 'school_attendance',
    'hh_edu_level': 'education_level',
    'hh_ag_activity__Crop': 'activity_crop',
    'hh_ag_activity__Livestock': 'activity_livestock',
    'hh_ag_activity__Poultry': 'activity_poultry',
    'hh_ag_activity__Fisheries': 'activity_fisheries',
    'hh_plots_number': 'num_plots',
    'hh_cereals__MAIZE': 'grow_maize',
    'hh_cereals__RICE': 'grow_rice',
    'hh_legumes__BEANS_COWPEA': 'grow_beans',
    'hh_tot_number__CATTLE': 'num_cattle',
    'hh_tot_number__GOAT': 'num_goat',
    'hh_tot_number__CHICKEN': 'num_chicken',
    'hh_fisheries_list__FISH_FARMING': 'fish_farming',
    'agri_estab': 'has_agri_establishment',
    'non_agri_estab': 'has_nonagri_establishment',
    'nasc_listing_weight': 'weight'
}
```

```python
df_filtered = df_sampled[list(columns_to_keep.keys())].rename(columns=columns_to_keep)

df_filtered.to_csv("FarmingHouseholds_filtered_renamed.csv", index=False)
```

```python
# Select and rename useful columns
selected_columns = {
    "household_id": "Household ID",
    "state": "State",
    "lga": "LGA",
    "hh_head_sex": "Gender",
    "hh_head_age": "Age",
    "school_attendance": "School Attendance",
    "education_level": "Education Level",
    "activity_crop": "Crop Farming",
    "activity_livestock": "Livestock Farming",
    "activity_poultry": "Poultry Farming",
    "activity_fisheries": "Fisheries",
    "num_plots": "Number of Plots",
    "grow_maize": "Grows Maize",
    "grow_rice": "Grows Rice",
    "grow_beans": "Grows Beans",
    "num_cattle": "Number of Cattle",
    "num_goat": "Number of Goats",
    "num_chicken": "Number of Chickens",
    "fish_farming": "Fish Farming",
    "has_agri_establishment": "Agri Establishment",
    "has_nonagri_establishment": "Non-Agri Establishment",
    "weight": "Survey Weight"
}

df = df_filtered[list(selected_columns.keys())].rename(columns=selected_columns)
df.head()
```
